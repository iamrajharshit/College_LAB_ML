```python
#extraction independem=nt and dependent variable
x=df.iloc[:,:-1].values
y=df.iloc[:,4].values
x
```

```python
#caltgorical value data
from sklearn.preprocessing import LabelEncoder
labelencoder_x= LabelEncoder()
x[0:,2]= labelencoder_x.fit_transform(x[0:,2])
x
```

```python
#Training and Testing Dataset Reg model
from sklearn.model_selection import train_test_split #it will split it 
x_train,x_test,y_train, y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train, y_train)
```

```python
pred=regressor.predict(x_test)
```
